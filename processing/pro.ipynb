```python
import pandas as pd
import glob
import os
import re
```

```python
# /color_dataのファイル名を取得
files = glob.glob('./color_data/*.csv')
print(files)
```

```python
# filesの1つ目のファイルを読み込み
for file in files:
    with open(file, 'r') as f:
        datas = f.read()
    datas = datas.split('\n')
    datas_copy = []
    for data in datas:
        data = str(data)
        data_list = data.split('\t')
        datas_copy.append(data_list)
    # df にデータを格納
    # dfの名前はfule から ./color_data/ を除いて .csv を除いたもの
    df_name = file.replace('./color_data/', '')
    df = pd.DataFrame(datas_copy[1:], columns=datas_copy[0])
    print(df)
    # dfをcsvで保存
    df.to_csv('./colors/' + df_name, index=False)
```

```python
def create_head(text):
    return '<h4>' + str(text) + '</h4>'
def create_div(text):
    if text != ' ':
        return '<div style="background-color: #' + str(text) + '; color: #' + str(text) + ';">'+ str(text) +'</div>'
    else:
        return '<div style="background-color: none; color: none;"></div>'
```

```python
# /color_dataのファイル名を取得
files = glob.glob('./colors/*.csv')
print(files)

# files を並べ替える
copy_files = []
# R → YR → Y → GY → G → BG → B → PB → P → RP の順番でその前に付く数字が小さい順に並べる
for file in files:
    copy_files.append(file.replace('./colors/', '').replace('.csv', ''))
copy_files.sort()
print(copy_files)
```

```python
# ./html/*/*.html を作る
# 入力された並び順をもとにhtmlを作る
def save_html(tmp,file_name):
    # ex imput: 7.5RP → num: 7.5 string: RP
    num_tmp = re.findall(r'\d+', file_name)
    # num_tmp の要素が1つのとき
    if len(num_tmp) == 1:
        num = num_tmp[0]
    # num_tmp の要素が2つのとき
    elif len(num_tmp) == 2:
        num = num_tmp[0] + '.' + num_tmp[1]
    else:
        print('error')
        return 1
    string = re.findall(r'[A-Z]+', file_name)[0]
    print(num, string)
    
    # ./html/string/num.html を作る
    # 必要なディレクトリはある
    with open('./html/' + string + '/' + num + '.html', 'w') as f:
        f.write(tmp)

    return 0
```

```python
# /color_dataのファイル名を取得
files = glob.glob('./colors/*.csv')
print(files)

html = ""
for file in files:
    file_name = file.replace('./colors/', '').replace('.csv', '')
    df = pd.read_csv(file)
    df_head = []
    df_head = df.columns.values.tolist()
    print(df_head)
    tmp = ""
    for i in range(len(df)):
        if i == 0:
            tmp += create_head(df_head[i])
            for n in range(14):
                # tmp += create_div(" ")
                tmp += create_head("c" + str(2 + n*2))
        for header in df_head:
            if header == file_name:
                tmp += create_head(df[header][i])
            else:
                tmp += create_div(df[header][i])
    # html += tmp
    save_html(tmp,file_name)
```
